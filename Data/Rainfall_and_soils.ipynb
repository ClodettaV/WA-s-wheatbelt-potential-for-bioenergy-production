# Import rainfall database from API endpoint

```python
url = "https://api.dpird.wa.gov.au/v2/weather/stations/rainfall?"

start_date = "2021-01-01"
end_date =  '2022-01-01'

response = requests.get(f"{url}startDate={start_date}&endDate={end_date}&offset=0&limit=1000&json&api_key={api_key}")

#https://api.dpird.wa.gov.au/v2/weather/stations/rainfall?startDate=2017-01-01&endDate=2017-01-03&stationCode=SP%2CAN001&offset=0&limit=25&includeClosed=false
```

```python
# Drop columns not needed
rainfall_df = df.drop(['links','rainfall.to9AM','rainfall.since9AM', 'rainfall.currentHour', 'rainfall.last24Hrs', 'rainfall.last7Days', 'rainfall.last14Days', 'rainfall.monthToDate'], axis=1)
rainfall_df
```

# Import Australian soils pH database from csv file

```python
# read csv file 
soil_df = pd.read_csv ('Australian soils pH.csv')
soil_df
```

```python
soil_df.columns = [x.lower() for x in soil_df.columns]
```

```python
soil_df
```

# Run shapely on both datasets

```python
from shapely.geometry import Point, Polygon

def apply_check_polygon(row):
    with open("lga.geojson", 'r') as f:
        lga_geojson = json.load(f)
    point = Point(row["longitude"], row["latitude"]) # shapely takes in x: lon, lat: y. 
    for feature in lga_geojson["features"]:
        shire_name = feature["properties"]["name"]
        if len(feature["geometry"]["coordinates"][0]) > 3:
            polygon = Polygon(feature["geometry"]["coordinates"][0])
            if polygon.contains(point):
                return shire_name 
    return "not found" 
```

```python
# run it for rainfall 
rainfall_df["shire"] = rainfall_df.apply(apply_check_polygon, axis=1)
rainfall_df[rainfall_df["shire"] != "not found"]
```

```python
# run it for soil 
soil_df["shire"] = soil_df.apply(apply_check_polygon, axis=1)
soil_df[soil_df["shire"] != "not found"]
```

```python
merged_df = pd.merge(soil_df, rainfall_df, on="shire", how="inner")
merged_df
```

```python
df_clean = merged_df[merged_df.shire != "not found"]
df_clean
```

```python
# Drop columns not needed
df_drop = df_clean.drop(['duplicate', 'bos_field_ph', 'rainfall.period'], axis=1)
df_drop
```

```python
df_drop['dateTime'] = df_drop["dateTime"].astype('datetime64').dt.date
df_drop
```

```python
# Remove ', SHIRE OF' from shire column
df_drop['shire'] = df_drop['shire'].str.replace(', SHIRE OF','')
df_drop
```

```python
df_drop['shire'] = df_drop['shire'].str.replace(', CITY OF','')
df_drop
```

```python
df_drop['shire'] = df_drop['shire'].str.lower()
df_drop
```

```python
df_rainAVG = df_drop.groupby('shire')[['rainfall.yearToDate','tos_field_ph']].mean().reset_index()
df_rainAVG
```
